```python
from office365.sharepoint.client_context import ClientContext
from office365.runtime.auth.user_credential import UserCredential
```

```python
SHAREPOINT_SITE = "https://mmiholdingsltd.sharepoint.com/:f:/s/HCEPeopleAnalyticsTeam/IgAtekbUlseSSqiQpvCU3RJcAS-1UI-dvOCqfD7pJyrYaG4?e=kjoHPt"
SHAREPOINT_DOC_LIBRARY = "Documents"
SHAREPOINT_FOLDER = "Automation"
SP_USERNAME = "svc-sharepoint@mmltd.co.za"
SP_PASSWORD = "********"
```

```python
def upload_to_sharepoint(local_file_path, target_file_name):
    """
    Uploads a file to a SharePoint folder
    """
    ctx = ClientContext(SHAREPOINT_SITE).with_credentials(
        UserCredential(SP_USERNAME, SP_PASSWORD)
    )

    target_folder = (
        ctx.web
        .get_folder_by_server_relative_url(
            f"/sites/PeopleAnalytics/{SHAREPOINT_DOC_LIBRARY}/{SHAREPOINT_FOLDER}"
        )
    )

    with open(local_file_path, "rb") as file:
        target_folder.upload_file(target_file_name, file.read()).execute_query()

    print(f"✅ Uploaded to SharePoint: {target_file_name}")
```
